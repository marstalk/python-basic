```python
import pandas as pd

sales_df = pd.read_csv('../dataset/sales_data.csv')
sales_df.head()
```

```python
# 1. overall statistics of number type columns
sales_df.describe()
```

```python
# 2. or we can statistics a specific column, .mean() .max() .min() 
# find the mean, max of profit of all products
sales_df.head()
profit_mean = sales_df["Profit"].mean()
profit_max = sales_df["Profit"].max()
print(f"mean profit is {profit_mean}")
print(f"max profit is {profit_max}")
```

```python
# 3. unique a column, return array
# list all regions
sales_df.head()
region_uniqe = sales_df["Region"].unique()
print(f"unique of region {region_uniqe}")
```

```python
# 3.1 values_count, count of each unique value
# find out the product count in different region
sales_df.head()
product_count_regions = sales_df["Region"].value_counts()
print(f"product counts by regions is {product_count_regions}")
```

```python
weather_df = pd.read_csv('../dataset/china_weather.csv', delimiter=';')
weather_df.head()
```

```python
# 4 协方差，超级厉害
# 两只股票，是不是同涨同跌？程度多大？正相关还是负相关？
# 产品销量的波动，跟哪些因素正相关、负相关，程度有多大？
# Do the two stocks rise and fall together? To what extent? Are they positively or negatively correlated?
# How does the fluctuation in product sales correlate with various factors? 
# Which factors have a positive or negative correlation, and to what extent?

# 协方差：衡量同向、反向程度，
#   如果协方差为正，说明同向变化，越大，同向程度越高。
#   如果协方差为负，说明反向变化，越小，反向程度越高。
# 英 /kəʊ'veərɪəns/
# Covariance: Measures the degree of simultaneous (same-direction) or opposite-direction changes.
#   If the covariance is positive, it indicates same-direction movement; the larger the value, the stronger the same-direction relationship.
#   If the covariance is negative, it indicates opposite-direction movement; the smaller the value, the stronger the opposite-direction relationship.

weather_df.cov()

```

```python
# specify columns
weather_df["Wind Speed  [10 m above gnd]"].cov(weather_df["Temperature  [2 m above gnd]"])
```

```python
# 4.1 相关系数，超级厉害
# 相关系数： 衡量相似度程度，-1到1之间，越接近1，正相关程度越高，越接近-1，负相关程度越高。
#   如果相关系数为0，说明两者没有线性相关性。
#   如果相关系数为1，说明两者完全正相关。
#   如果相关系数为-1，说明两者完全负相关。
# 英 /ˌkɒrə'leɪʃ(ə)n/
# Correlation Coefficient: Measures the degree of similarity, ranging from -1 to 1. The closer it is to 1, the stronger the positive correlation; the closer it is to -1, the stronger the negative correlation.
#   A correlation coefficient of 0 means no linear correlation between the two variables.
#   A correlation coefficient of 1 means a perfect positive correlation.
#   A correlation coefficient of -1 means a perfect negative correlation.
weather_df.corr()
```

```python
# specify columns
weather_df["Wind Speed  [10 m above gnd]"].corr(weather_df["Temperature  [2 m above gnd]"])
```
